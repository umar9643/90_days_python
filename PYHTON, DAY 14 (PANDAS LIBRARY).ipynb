PANDAS IS A PYTHON LIBRARY USED TO WORK WITH DATASET

```python
IT HAS FUNCTION FOR ANALYZING, CLEANING, EXPLORING, AND MANUPLATING DATA
```

```python
THE NAME PANDAS HAS A REFENRECE FROM BOTH "PANEL DATA", "PYHTON DATA ANALYSIS" WAS CREATED BY WES MCKINNEY IN 2008
```

```python
IT CAN CLEAN MESSY DATA SETS AND MAKE THEM REDABLE AND REVEVENT
```

```python
RELEVENT DATA IS VERY INPORTANT IN DATA SCIENCE
```

# ADVANTAGES OF USING PANDAS

FAST AND EFFICEINT FOR MANUPLATING AND ANALYZING DATA
DATA FROM DIFFERENT FILE AND OBJECT CAN BE LOADED
EASY HANDLING OF MISSING DATA(REPRESENTED AS NAN)IN FLOATING POINTS AND AS WELL AS NON FLOATING POINTS DATA
SIZE MUTABILITY: COLUMN CAN BE INSERTED AND DELETED FROM DATAFRAME AND HIGHER DIMENTION OBJECTS
DATASET MERGING AND JOINING
FLEXIBLE RESHAPING AND PIVOTING OF DATASES
PROVIDES TIME SERIES FUNCTIONALITY
POWERFULL GROUPBY FUNCTIONALITY FOR PERFORMINGSPLIT APPLY-COMBINE OPERATION ON DATASET
