1.SERIES: IT IS DEFINES AS ONE DIMENTIONAL ARRAY THAT IS CAPABLE OF STORING VARIOES DATA TYPES'

```python
import  pandas as pd
```

```python
x=[1,2,3]
var=pd.Series(x)
print(var)
print(type(var))
print(var[2])
```

```python
x=[1,2,3]
var=pd.Series(x,index=["one","two","three"])#adding index here
print(var)
print(type(var))
print(var[2])
```

```python
x=[1,2,3]
var=pd.Series(x,index=["one","two","three"],dtype="float")#changing datatypes here
print(var)
print(type(var))
print(var[2])
```

```python
x=[1,2,3]
var=pd.Series(x,index=["one","two","three"],name="umar")#adding name here
print(var)
print(type(var))
print(var[2])
```

```python
#making series from dictionary
dic={"name":["umar","imran","ansh"],"marks":[8.22,6.8,7.9],"rank":[1,3,2]}
dict=pd.Series(dic)
dict
```

```python
s=pd.Series(14)#making series from only one elemnet
print(s)
print(type(s))
```

```python
s=pd.Series(14,index=[1,2,3,4,5])#making series from only one elemnet 
print(s)
print(type(s))
```

# HANDLING MISSING DATA(DROPNA AND FILLNA)

```python
import pandas as pd 
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
data
```

# DROPNA

```python
data.dropna()
```

```python
data
```

```python
data.dropna(axis=0)#work along rows
```

```python
data.dropna(axis=1)#works along columns
```

```python
data
```

```python
data.dropna(subset=["Cabin"])
```

```python
data.dropna(thresh=1)
```

# FILLNA

```python
data
```

```python
#to fill data in blank 
data.fillna("umar")
```

```python
# to fill diff data in diff col
data.fillna({"Cabin":"umar","Age":"420"})
```

```python
#forward filling
data.fillna(method="ffill")
```

```python
#backward filling
data.fillna(method="bfill")
#we can also use this fn along axis 
```

```python
#inplace = True for permanent chnage
```

# INTERPOLATE

```python
import pandas as pd
```

```python
var=pd.read_csv("https://raw.githubusercontent.com//datasciencedojo//datasets//master//titanic.csv")
var
```

```python
#interpolate is used to fill nan by their own
#works for numerical data only
var.interpolate()
```

method:{linear,time,index,values,nearest,zero,slinear,quadratic,barycentric,krogh,polynomial,
       spline,piecewise_polynomial,from_derivatives,pchip,akima}

```python
var.interpolate(method="zero")
```

```python
var.interpolate(method="index")
```

```python
var.interpolate(method="index",limit=2)
```

```python
var.interpolate(limit_direction="forward",limit=2)
```

```python
var.interpolate(limit_direction="backward",limit=2)
```

```python

var.interpolate(limit_direction="both",limit=2)
```
