```python
#importing matplotlib moduole
from matplotlib import pyplot as plt

#x-axis values
x=[2,3,5,4,6]

#y-axis values
y=[3,4,5,1,2]

#function to plot
plt.plot(x,y, )

#function to show the plot
plt.show()
```

```python
#x-axis values
x=[2,3,5,4,6]

#y-axis values
y=[3,4,5,1,2]

#function to plot
plt.plot(x,y,marker="o" )#for adding marker on edges use marker="o"

#function to show the plot
plt.show()
```

```python
#x-axis values
x=[4,3,4,5,6]

#y-axis values
y=[4,5,6,2,3]

plt.plot(x,y)#for plotting x,y
plt.plot(y,marker = "*")#for plotting only y
plt.show()
```

```python
x=[1,2,3,4,5]
y=[5,6,7,8,9]
plt.bar(x,y, color="g")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[5,6,7,8,9]
z=["r","g","b"]
plt.bar(x,y, color=z)
plt.show()
```
