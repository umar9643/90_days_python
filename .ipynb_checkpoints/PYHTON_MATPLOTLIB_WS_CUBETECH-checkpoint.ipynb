# PIE CHART

normal basic syntax

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
plt.pie(x)
plt.show()
```

adding label

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
plt.pie(x,labels=y)
plt.show()
#labels is used to give label
```

explore parameter

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
ex=[.4,.0,.0,.0,]
plt.pie(x,labels=y,explode=ex)
plt.show()
#labels is used to give label
#explode is used to take out a slide
```

changing fefault colors

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
ex=[.4,.0,.0,.0,]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
#labels is used to give label
#explode is used to take out a slide
#colors is used to chnage colors of slices
```

showing persentages

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
ex=[.4,.0,.0,.0,]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%.1f%%")
plt.show()
#labels is used to give label
#explode is used to take out a slide
#colors is used to chnage colors of slices
#used autopct="%.1f%%" for showing persentage
```

shadow parameter

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
ex=[.4,.0,.0,.0,]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%.1f%%",shadow="True")
plt.show()
#labels is used to give label
#explode is used to take out a slide
#colors is used to chnage colors of slices
#used autopct="%.1f%%" for showing persentage
#used shadow parameter
```

radius parameter

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
ex=[.4,.0,.0,.0,]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%.1f%%",shadow="True",radius=1.5)
plt.show()
#labels is used to give label
#explode is used to take out a slide
#colors is used to chnage colors of slices
#used autopct="%.1f%%" for showing persentage
#used shadow parameter
#radius parametr for adjusting the size of circle
```

labeldistance parameter

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
ex=[.4,.0,.0,.0,]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%.1f%%",shadow="True",radius=1.5,labeldistance=.9)
plt.show()
#labels is used to give label
#explode is used to take out a slide
#colors is used to chnage colors of slices
#used autopct="%.1f%%" for showing persentage
#used shadow parameter
#radius parametr for adjusting the size of circle
#labeldistance parameter for deciding the label distance from plot
```

startangle

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
ex=[.4,.0,.0,.0,]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%.1f%%",shadow="True",radius=1.5,labeldistance=.9,startangle=90)
plt.show()
#labels is used to give label
#explode is used to take out a slide
#colors is used to chnage colors of slices
#used autopct="%.1f%%" for showing persentage
#used shadow parameter
#radius parametr for adjusting the size of circle
#labeldistance parameter for deciding the label distance from plot
#startangle parameter for agle adjustment
```

textprops

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["a","b","c","d"]
ex=[.4,.0,.0,.0,]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%.1f%%",shadow="True",radius=1,labeldistance=.9,startangle=90,
       textprops={"fontsize":15})
plt.show()
#labels is used to give label
#explode is used to take out a slide
#colors is used to chnage colors of slices
#used autopct="%.1f%%" for showing persentage
#used shadow parameter
#radius parametr for adjusting the size of circle
#labeldistance parameter for deciding the label distance from plot
#startangle parameter for agle adjustment
#textprops properties of text
```

rotatelebels

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["abh","bhbhj","jghvc","jhbd"]
ex=[.4,.0,.0,.0,]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%.1f%%",shadow="True",radius=1,labeldistance=.9,startangle=90,
       textprops={"fontsize":15},rotatelabels="True")
plt.show()
#labels is used to give label
#explode is used to take out a slide
#colors is used to chnage colors of slices
#used autopct="%.1f%%" for showing persentage
#used shadow parameter
#radius parametr for adjusting the size of circle
#labeldistance parameter for deciding the label distance from plot
#startangle parameter for agle adjustment
#textprops properties of text
#rotatelabels is used for rotating labels
```

adding legend

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["abh","bhbhj","jghvc","jhbd"]
ex=[.4,.0,.0,.0,]
c=["r","b","g","y"]


plt.pie(x,labels=y,explode=ex,colors=c,autopct="%.1f%%",shadow="True",radius=.7,labeldistance=.9,startangle=90,
       textprops={"fontsize":15},rotatelabels="True")
plt.legend(loc=2)
plt.show()
#labels is used to give label
#explode is used to take out a slide
#colors is used to chnage colors of slices
#used autopct="%.1f%%" for showing persentage
#used shadow parameter
#radius parametr for adjusting the size of circle
#labeldistance parameter for deciding the label distance from plot
#startangle parameter for agle adjustment
#textprops properties of text
#rotatelabels is used for rotating labels
#plt.legend(loc=n) for adding legend
```

# STACK PLOT

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area=[3,2,5,4,6]

plt.stackplot(x,area)
plt.show()
#basic sing area chart
```

```python
x=[1,2,3,4,5]
area=[3,2,5,4,6]
area1=[3,2,6,4,7]
area2=[1,2,5,3,6]

plt.stackplot(x,area,area1,area2)
plt.show()
#basic sing area chart
#basic multiplt area chart
```

```python
x=[1,2,3,4,5]
area=[3,2,5,4,6]
area1=[3,2,6,4,7]
area2=[1,2,5,3,6]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l)
plt.legend()
plt.show()
#basic sing area chart
#basic multiplt area chart
#plt.legends to add legends note it works with lables = l
```

```python
x=[1,2,3,4,5]
area=[3,2,5,4,6]
area1=[3,2,6,4,7]
area2=[1,2,5,3,6]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","b","y"])
plt.legend()
plt.show()
#basic sing area chart
#basic multiplt area chart
#plt.legends to add legends note it works with lables = l
#colors parametrs to change default colors
```

```python
x=[1,2,3,4,5]
area=[3,2,5,4,6]
area1=[3,2,6,4,7]
area2=[1,2,5,3,6]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","b","y"])
plt.legend()
plt.show()
#basic sing area chart
#basic multiplt area chart
#plt.legends to add legends note it works with lables = l
#colors parametrs to change default colors
```

```python
x=[1,2,3,4,5]
area=[3,2,5,4,6]
area1=[3,2,6,4,7]
area2=[1,2,5,3,6]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","b","y"],baseline="sym")
plt.legend()
plt.show()
#basic sing area chart
#basic multiplt area chart
#plt.legends to add legends note it works with lables = l
#colors parametrs to change default colors
#baseline parameters for adjusting base["zero","sym","wiggle"]
```

# STEP PLOT

```python
x=[1,2,3,4,5,6]
y=[2,3,4,5,7,9]
plt.step(x,y)
plt.show()
#basic symtax of step plot
```

```python
x=[1,2,3,4,5,6]
y=[2,3,4,5,7,9]
plt.step(x,y)
plt.grid()
plt.show()
#basic symtax of step plot
#plt.grid parameter for adding grid
```

```python
x=[1,2,3,4,5,6]
y=[2,3,4,5,7,9]
plt.step(x,y)

plt.title("step plot")
plt.xlabel("xxxxx")
plt.ylabel("yyyyy")
plt.show()
#basic symtax of step plot
#plt.grid
#adding title,xlabel,ylabel
```

```python
x=[1,2,3,4,5,6]
y=[2,3,4,5,7,9]
plt.step(x,y,color="r",marker="*",markersize=11)

plt.title("step plot")
plt.xlabel("xxxxx")
plt.ylabel("yyyyy")
plt.show()
#basic symtax of step plot
#plt.grid
#adding title,xlabel,ylabel
#using color,marker,markersize parameters
```
