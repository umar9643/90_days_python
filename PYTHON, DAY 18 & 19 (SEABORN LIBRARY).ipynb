# BAR PLOT IN SEABORN LIBRARY

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=sns.load_dataset("penguins")
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x="island",y="bill_length_mm", data=df1)
plt.show()
```

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm,hue=df1.sex)
```

```python
df1
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex")
```

# ORDER OF X AXIS

```python
order1=["Torgersen","Biscoe","Dream"]
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex",order=order1)
plt.show()
```

```python
order1=["Torgersen","Biscoe","Dream"]
ho1=["Female","Male"]
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex",order=order1,hue_order=ho1)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,color="green",hue="sex")
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex",palette="rocket_r")
plt.show()
```

# SATURATION

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,saturation=1)
```

# HISTOGRAM

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1
```

```python
sns.displot(df1["flipper_length_mm"])#displot in seaborn and histplot in matplotlib
plt.show()
```

# BINS

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230])
plt.show()
```

# COLOR

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230],color="green")
```

# KDE(KERNAL DENSITY ESTIMATOR)

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230],color="red" ,kde=True)
```
