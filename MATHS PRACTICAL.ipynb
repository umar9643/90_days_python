```python
mat= np.matrix( [[5,4,3],
                [9,7,8],
                [3,2,7]] )
print(mat)
z=np.linalg.det(mat)
print("\nDeterminant:", int(z))
inverse=np.linalg.inv(mat)
print("Inverse of the matrix is:\n", inverse)
print("The trace of the given matrix is:\, np.trace(mat)")
print("The transpose of the given matrix is:\n", np.transpose(mat))
```

```python
def compute_eigen(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues, eigenvectors

matrix = np.array([[6, 7],
                   [8, 9]])

eigenvalues, eigenvectors = compute_eigen(matrix)

print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values
x = np.linspace(-2*np.pi, 2*np.pi, 400)

# Compute y values for trigonometric functions
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# Plot the trigonometric functions
plt.figure(figsize=(10, 6))

# Plot sine function
plt.plot(x, y_sin, label='Sin(x)', color='blue')

# Plot cosine function
plt.plot(x, y_cos, label='Cos(x)', color='red')

# Plot tangent function (with limited y-axis for visibility)
plt.plot(x, y_tan, label='Tan(x)', color='green')
plt.ylim(-10, 10)

# Set plot title and labels
plt.title('Trigonometric Functions')
plt.xlabel('x')
plt.ylabel('y')

# Add grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the polynomial coefficients
coefficients = [1, -9, -4, 7, 1]  # Represents: x^4 - 4x^2 - 3x + 1

# Create an array of x values
x = np.linspace(-5, 5, 400)

# Compute the y values using numpy's polyval function
y = np.polyval(coefficients, x)

# Plot the polynomial
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^4 - 4x^2 - 3x + 1', color='blue')
plt.title('Polynomial Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the linear function parameters
m = 2  # Slope
b = 3  # Intercept

# Create an array of x values
x = np.linspace(-5, 5, 400)

# Compute y values for the linear function
y = m * x + b

# Plot the linear function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'y = {m}x + {b}', color='blue')

# Set plot title and labels
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')

# Add grid and axes lines
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Display legend and show plot
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the exponential function parameters
a = 2  # Base
b = 0.5  # Exponent
c = 1  # Vertical shift

# Create an array of x values
x = np.linspace(-2, 2, 400)

# Compute y values for the exponential function
y = a * np.exp(b * x) + c

# Plot the exponential function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'y = {a} * e^({b}x) + {c}', color='blue')

# Set plot title and labels
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')

# Add grid and axes lines
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Display legend and show plot
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values (excluding zero for logarithm)
x_log = np.linspace(0.1, 5, 400)

# Compute y values for the logarithmic function (base "e")
y_log = np.log(x_log)

# Plot the logarithmic function
plt.figure(figsize=(8, 6))
plt.plot(x_log, y_log, label='y = ln(x)', color='blue')

# Set plot title and labels
plt.title('Logarithmic Function (Base "e")')
plt.xlabel('x')
plt.ylabel('y')

# Add grid and axes lines
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Display legend and show plot
plt.legend()
plt.show()
```

```python
# Create an array of x values (excluding negative values for square root)
x_sqrt = np.linspace(0, 5, 400)

# Compute y values for the square root function
y_sqrt = np.sqrt(x_sqrt)

# Plot the square root function
plt.figure(figsize=(8, 6))
plt.plot(x_sqrt, y_sqrt, label='y = sqrt(x)', color='green')

# Set plot title and labels
plt.title('Square Root Function')
plt.xlabel('x')
plt.ylabel('y')

# Add grid and axes lines
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Display legend and show plot
plt.legend()
plt.show()
```
